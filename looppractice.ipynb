```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# print number from 1 to 10

for i in range(1, 11):
    print(i, end=" ")
```

```python
# print even numbers from 1 to 20 in one line

for i in range(2,21,2):
    print(i, end=",")
```

```python
for i in range(1,21):
    if i % 2 == 0:
        print(i, end=", ")
```

```python
num = int(input("Enter a number: "))

for i in range(1,11):
    print(num, "x", i, "=", num*i)
```

```python
for i in range(3):        # outer loop
    for j in range(3):    # inner loop
        print("*")
```

```python
for i in range(1,6):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
for i in range(5,0,-1):
    for j in range(i):
        print("*", end=" ")
    print()
```
