```python
institute = "Edukron"
tech_stack = "Azure DevOps and Python"
```

```python
first_character = institute[0]
print(f"First character of '{institute}' is {first_character}")
```

```python
last_character = institute[-1]
print(f"The last character of '{institute}' is {last_character}")
```

```python
#slicing
first_5 = tech_stack[0:5]
print(f"The first 5 character of sliced string is '{first_5}'")
```

```python
last_character = tech_stack[-6:]
print(f"the last 6 character of sliced string is '{last_character}'")
```

```python
#get everything from institute except the first and last character
middle_characters = institute[1:-1]
print(f"The middle characters except first and last characters is '{middle_characters}'")
```

```python
every_second = institute[::2]
print(f"Every second character of '{institute}' is '{every_second}'")
```

```python
#Reverse the string by using slicing

reversed_institute = institute[::-1]
```

```python
print(f"The reversed of institute character is '{reversed_institute}'")
```

```python
#Get a slice from  index 6 onwards

from_index_6 = tech_stack[6:]
print(f"The sliced from the tech_stack from index 6 is '{from_index_6}'")
```
