list.insert(index,item)

Insert an item at a specific index. Existing items from the index onwards are shifted to the right. modifies the list in-place.

```python
course_list = ["Data science", "Data engineering", "Machine learning"]
priority_tasks = ["deploy App", "Monitor logs"]
```

```python
course_list.insert(0, 'Python Basics')
print(f"The new course added in the first position is: {course_list}")
```

```python
tasks = ["Task A", "Task B"]
tasks.insert(len(tasks), "Task C")
print(f" The new task is added at the end of task list: {tasks}")
```

```python
#insert list as a item 

nested = [1, [3, 4]]
nested.insert(1,2)
print(f"the new list after inserting is: {nested}")
```

Membership Testing (in, not in)

Checks if an item exists within the list. Returns True or False.

```python
courses = ["Data science", "Data engineering", "Machine learning", "Python basics"]
```

```python
is_ML_offered = "Machine learning" in courses
print(f"The course is offered in the list: {is_ML_offered}")
```

```python
check_courses = "Data science" in courses
print(f"the course is not listed in the courses: {check_courses}")
```
