String lower and String upper -- str.lower() and str.upper()

```python
institute = "Edukron"
focus_area = "Data Science"
tech_stack = "Azure DevOps and Python"
```

```python
#institute name in lower case

lower_institute = institute.lower()
print(f"The institute name is in lower case is '{lower_institute}'")
```

```python
upper_focus = focus_area.upper()
print(f" The upper case of focus area is '{upper_focus}'")
```

```python
lower_stack = tech_stack.lower()
print(f"The lower case of tech stack is '{lower_stack}'")
```

```python
mixed_case = "PyThOn"
upper_mixed = mixed_case.upper()
print(f"The upper case of mixd character is '{upper_mixed}'")
```

```python
#trying 1st character in upper case and rest in lower case

title = "pyThon ProGramming"
title_case = title.title()
print(f"The title case of the string is '{title_case}'")
```

str.strip(), str.lstrip(), str.rstrip()

```python
user_input = "Data Science"
cleaned_inputs = user_input.strip()
print(f" Striped user input is '{cleaned_inputs}'")
```

```python
user_input1 = "  Data Science  "
cleaned_inputs1 = user_input1.strip()
print(f"After stripping of the user input is '{cleaned_inputs1}'")
```

```python
#Removing leading space

leading_space = "   Data Science"
cleaned_leading = leading_space.lstrip()
print(f"The leading space is removed from '{leading_space}' and claeded string is '{cleaned_leading}'")
```

```python
#stripping of specific character from the string

course_code = "+++DS101==="
stripped_code = course_code.strip("+=")
stripped_code
```

```python
path = "course.pdf"
file_extension = path.split(".")[-1]
print(f"The file extension of the path is '{file_extension}'")
```

```python
version = "v1.0.Release..." 
stripped_version = version.rstrip(".Release")
print(f"The stripped version is '{stripped_version}'")
```

```python
version = "v1.0.Release..."
```

```python
stripped_version = version.rstrip(".")
print(f"The stripped version is '{stripped_version}'")
```

```python
version = "v1.0.RELEASE..."
stripped_version = version.rstrip(".RELAS") # Removes '.', 'R', 'E', 'L', 'A', 'S' from end
print(f"7. Right-stripped specific chars: '{stripped_version}'")
```
