## Scope of variable

Local scope..

- Variable defined inside a function..

```python
def my_function():
    local_var = 10 #Local variable.
    print(f"Inside function: {local_var}")
```

```python
my_function()
```

## Global scope..
- Variable defined outside function

```python
global_var = 100 #Global variable

def my_function():
    print(f"Inside function: {global_var}")
```

```python
my_function()
```

```python
global_var
```

```python
## local vs global with same name

x = 50

def my_function():
    x = 10
    print(f"Inside function: {x}")
```

```python
my_function()
```

```python
print(f"Outside function: {x}")
```

```python
# Modifying Global variable

count = 0  #global variable

def increment():
    global count   # Declare thet we want to modify global variable
    count += 1
    print("Count: ", count)
```

```python
increment()
increment()
print(f"Count outside: {count}")
```

### Lambda Function(Anonymous Functions)

A lambda function is a small anonymous function defined in one line..

#### Syntax = lambda arguments:expression

```python
#regular function

def add(a,b):
    return a + b
```

```python
## Lambda function for same output

add_lambda = lambda a, b: a + b
```

```python
print(add(5,3))
print(add_lambda(5,10))
```

```python
## Lambda with single Argument:

square = lambda x: x**2
print(square(6))
```

```python
# Lambda with multiple arguments

area = lambda length, width: length * width
```

```python
print(area(88,34))
```

```python
## Lambda with condition

# check is even   # lambda arg: expression if condition else expression 2

is_even = lambda x: "Even" if x % 2 == 0 else "odd"
```

```python
is_even(49)
```

```python
is_even(100)
```

```python
# lambda arg: expression if condition else expression2

# Positive/negative

is_positive = lambda x: "Positive" if (x > 0) else "Negative"
```

```python
is_positive(23)
```

```python
is_positive(-98)
```

```python
# Pass/Fail (pass marks >= 50)

is_pass = lambda x: "Pass" if x>=50 else "Fail"
```

```python
is_pass(75)
```

```python
is_pass(49)
```

### Using lambda with built in functions

     1. map() -
     - Apply a function to every element in a list(or iterable)
syntax: map(function, iterable) returns: transformed values
numbers = [10,20,30,40]
"Takes each item -> apply logic -> return new list"

```python
## without lambda

def square(x):
    return x * x

numbers = [1,2,3,4,5]

result = list(map(square, numbers))
```

```python
result
```

```python
numbers = [1,2,3,2,4,6]
result2 = list(map(lambda x: x*2,numbers))
result2
```

### filter()

- Select only elements that satisfy a condition
- syntax: filter(function, iterable)
- returns only matching itmes
- "check condition -> keep only true value"

```python
numbers = [1,2,3,4,5]

result = list(filter(lambda x: x % 2==0, numbers))
result
```

### sorted()

- sorts elements
- sorted(iterable, key=function)

key defines hoe to sort...

```python
## very important interview perspective question

students = [("ram",80), ("sham",95), ("denny", 70)]
```

```python
students
```

```python
result = sorted(students)
```

```python
result1 = sorted(students, key= lambda x: x[1])
```

```python
result1
```

```python
students[0][1]
```

```python
employees = [
    ("John", 50000, 5),  #(name, salary, experience)
    ("Alice", 70000, 3),
    ("Bob", 50000, 7)
]
```

```python
# sort by salary, then experience

emp_salary = sorted(employees, key = lambda x: x[1])
emp_salary
```

```python
emp_experience = sorted(employees, key = lambda x:x[2])
emp_experience
```

```python
products = [
    ("boult", 65000, 9),
    ("Noise", 52000, 6),
    ("Samsung", 59000, 8),
    ("Nokia", 67000, 6),
    ("OnePlus", 55000, 7),
]
```

```python
filtered = list(filter(lambda x: x[1]<60000 and x[2]>4.5,products))
result = list(map(lambda x: (
    x[0],
    x[1],
    x[2],
      "Excellent" if x[2] >= 8 else "Good" if x[2] >= 7 else "Average"), filtered))
print(result)
```
