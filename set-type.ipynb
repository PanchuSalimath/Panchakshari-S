## Set Types

Set types represents unordered collection of unique items. They are useful for membership testing, removing dublicates, an performing mathematical set operation (like Union, intersection, difference).

# set 

Sets are mutable(changeble), unordered collections of unique, hashable items. Hashable means the items must be immutable (like number, strings, tuples, but not lists or other sets). sets are defined using curly braces {} or the set() constructor. An an empty set must be created with set() , as {} creates an empty dictionary.

```python
# 1. Creation and add(item)

data_analytics_topics = {"Visualization", "Power BI", "EDA", "Reporting", "Power BI"}
print(f"Analytical topic set: {data_analytics_topics}")
```

```python
# 2. Create a set using set() from a list (removes duplicates)
course_list = ["Data Science", "Data Engineering", "Azzure DevOps", "Data Science"]
unique_courses = set(course_list)
print(f"Unique courses: {unique_courses}")
```

```python
# 3. Create an empty set

required_skills = set()
print(f"Empty set: {required_skills}")
```

```python
# 4. Add a skill to empty set

required_skills.add("python")
print(f"The new required skill is: {required_skills}")
```

```python
# 5. Add another skill to the set

required_skills.add("SQL")
print(f" Updated second skill in required skillls: {required_skills}")
```

```python
# 6. Add an existing skill

required_skills.add("python")
print(f"Added an existing skill: {required_skills}")
```

```python
# 7. Add a number to set

required_skills.add(123)
required_skills
```

```python
# 8. Add a number to set

batch_ids = {101,102,104}
batch_ids.add(103)
batch_ids
```

```python
# 9. Add a tuple(hashable) to a set

config_params = {("host", "localhost")}
config_params.add(("port",8080))
print(f"Config params set: {config_params}")
```

```python
# 10. Create set from a string (unique characters)

unique_char = set("Edukron Institute")
print(f"Unique char in 'Edukron Institute': {unique_char}")
```

## update(iterable)
Explanation: Adds all unique items from an iterable (like a list, tuple, set, or string) into the set. Modifies the set in-place. Equivalent to set1 |= set2 | set3 ...

```python
devops_tools = {"Docker", "Kubernetes"}
data_tools = {"Python", "SQL", "Pandas"}
new_tools_list = ["Terraform", "GitHub Actions", "Docker"]
more_skills = ("Git", "SQL")
```
