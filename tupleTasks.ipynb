```python
#task 1: Add your fevorate food

foods = ["pizza", "bueger", "pasta"]
foods.append("Veg Biriyani")
print(f"My fevorate foods are: {foods}")
```

```python
#Task 2:Add multiple items

colors = ["red", "blue"]
colors.extend(["black", "white", "yellow"])
print(f"My fevorate colors are: {colors}")
```

```python
#Task 3: Insert at specific position

numbers = [1, 2, 4, 5]
numbers.insert(2,3)
print(f"Numbers are: {numbers}")
```

```python
#Task 4: Remove an item

fruits = ["apple", "banana", "mango", "grapes"]
fruits.remove("mango")
print(f"Fruits are: {fruits}")
```

```python
#Task 5: Remove last item

items = ["pen", "book", "eraser"]
items.pop(-1)
print(f"Items: {items}")
```

```python
#Task 6: Find Index

animals = ["cat", "dog", "lion", "tiger"]
index = animals.index("lion")
print(f"The index of lion is: {index}")
```

```python
#Task 7: Count Occurences

nums = [1, 2, 2, 3, 2, 4]
count =nums.count(2)
print(f" The number of 2s in the list is: {count}")
```

```python
#Task 8: Sort the list

marks = [85, 45, 90, 60]
marks.sort()
print(f"The sorted list of marks is: {marks}")
```

```python
#Task 9: Revers the list

letters = ["a", "b", "c", "d"]
letters.reverse()
print(f"The reversed list of letters is: {letters}")
```

```python
#Task 10: Copy the list

original = [10, 20, 30]
copied_list = original.copy()
print(f"The copied list from the original list is: {copied_list}")
```

```python
# Task 11: Clear the list
data = [1, 2, 3, 4]
data.clear()
print(f"After clearing the list from the data is: {data}")
```
