# Functions (user defined function)

- built-in (comes with python installation, such len(), print(), sum())

A function is a reusable block of code that performs a specific task. Instead of writing the same code multiple
times, you write it once in a function and call it whenever needed.

```python
## calculate avg of three different groups...

group1_sum = 85 + 90 + 78
group1_avg = group1_sum/3
print(f"Group 1 Average: {group1_avg}")
```

```python
group2_sum = 78 + 98 +35
group2_avg = group2_sum / 3
print(f"Group 2 Average: {group2_avg}")
```

```python
group3_sum = 12 + 32 + 34
group3_avg = group3_sum / 3
print(f"Group 3 Average: {group3_avg}")
```

```python
## same logic with function
def calculate_average(scores):
    """
    This function takes list of numbers and returns the average
    """
    total = sum(scores)
    avg = total/len(scores)
    return avg

print(calculate_average([12,23,22]))
print(calculate_average([525,322,33]))
print(calculate_average([65.44,34.54,344.54]))
```

def : keyword to define a function..

function_name = Name of the fucntion(follow variable name rules)

paramters : Input values (optional)

docstring = Documentation string(optional but recomended)

Function_body : Code that executes when function is called..

return : Sends result back to caller (optional)

```python
def greet():
    """Prints a greeting message"""
    print("Hello! Welcome to python function")
```

```python
# call the function
greet()
```

```python
## function with parameter

def greet_person(name):
    """Greets a person by name"""
    print(f"Hello, {name}! welcome to function")
```

```python
## calling function with parameter
name = input("Enter your name here: ")
greet_person(name)
```

```python
# Function with return value

def add_number(a,b):
    result = a+b
    return result
```

```python
add_number(7,87)
```

```python
greeting = greet_person("amar")
```

## Types of function arguments..
-Positional Arguments..
Arguments passed in the correct order


```python
def calculate_bmi(weight, hight):
    """Calculate Body Mass Index"""
    bmi = weight / (hight** 2)
    return bmi
```

```python
bmi1 = calculate_bmi(53 , 1.5)
bmi1
```

```python
## Keyword arguments
# Arguments passed with parameter name (order doesn't matter)

def student_info(name, age, grade):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"Grade: {grade}")
```

```python
student_info(name = "Panchakshari", age = 28, grade = "A")
```

```python
student_info(name = "Panchakshari")
```

## Default Argument

-Parameters with default values.

```python
def greet(name , message = "Hello"):
    print(f"{message}, {name}")
```

```python
greet("Bob")
```
