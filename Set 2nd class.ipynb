## remove(item) vs discard(item)

Both remove an item from the set.

remove(item): Removes item. Raises a KeyError if item is not found.
discard(item): Removes item if present. Does not raise an error if item is not found. Both modify the set in-place.

```python
devops_tools = {"Docker", "Kubernetes", "Terraform", "GitHub Actions"}
```

```python
devops_tools.remove("Terraform")
print(f"Set after removing 'Terraform': {devops_tools}")
```

```python
devops_tools.remove("Terraform")
print(devops_tools)
```

```python
# discard "Kubernetes" using discard() method

devops_tools.discard("Kubernetes")
print(f"Set after discarding 'Kubernetes': {devops_tools}")
```

```python
# Attempt to discard "kubernetes" again

devops_tools.discard("Kubernetes")
print(devops_tools)
```

## pop()

Removes and returns an arbitrary element from the set. Raises a KeyError if the set is empty. Since set are unorderd, you cannot predict which element will be popped. Modifies the set in-place.

```python
data_analytics_topics = {"Visualization", "Power BI", "EDA", "Reporting"}
numbers = {101, 102, 103}
```

```python
# Pop an orbitrary topic

popped_topic = data_analytics_topics.pop()
print(f"The popped topic is: {popped_topic}")
```

```python
popped_number = numbers.pop()
print(f"The popped number is: {popped_number}")
```

```python
print(data_analytics_topics)
print(numbers)
```

### Set Operations (Union, Intersection, Difference, Symmetric Difference)
Explanation: Mathematical operations on sets.

_Union (| or union()): Returns a new set with all items from both sets._

_Intersection (& or intersection()): Returns a new set with items common to both sets._

_Difference (- or difference()): Returns a new set with items in the first set but not in the second._

_Symmetric Difference (^ or symmetric_difference()): Returns a new set with items in either set, but not in both. In-place versions exist for some (e.g., update() is like union, intersection_update(), difference_update(), symmetric_difference_update())._

```python
data_sci_skills = {"Python", "Pandas", "ML", "SQL"}
data_eng_skills = {"Python", "SQL", "ADF", " ADLS", "Databricks"}
devops_skills = {"GitHub", "Pipelines", "Terraform", "Python"}
```

```python
all_data_skills = data_sci_skills.union(data_eng_skills)
print(f"The union of data science and data engineering skills is: {all_data_skills}")
```

```python
common_data_skills = data_sci_skills.intersection(data_eng_skills)
print(f"Intersection (DS & DE): {common_data_skills}")
```

```python
# skills unique to data science (DS - DE)

ds_only_skills = data_sci_skills.difference(data_eng_skills)
print(f"Intersection (DS - DE): {ds_only_skills}")
```

```python
# Skills unique to data engineering (DE - DS)

de_only_skills = data_eng_skills.difference(data_sci_skills)
print(f"Difference (DE - DS): {de_only_skills}")
```

```python
# symmetric difference (skills in DS, but not both)

non_common_data_skills = data_sci_skills.symmetric_difference(data_eng_skills)
print(f"symmetric difference (DE - DS): {non_common_data_skills}")
```

```python
common_all_three = data_sci_skills & data_eng_skills & devops_skills
print(f"6. Intersection (DS & DE & DevOps): {common_all_three}")
```

```python
ds_not_devops = data_sci_skills - devops_skills
print(f"7. Difference (DS - DevOps): {ds_not_devops}")
```
